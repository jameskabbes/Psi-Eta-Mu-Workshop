### Loop through, find Friday the 13ths

```python
total = 0

current_date = start_date
while current_date <= end_date:

    if current_date.day == 13: #13th
        if current_date.weekday() == 4: #Friday
            total += 1
            print (current_date)

    current_date = current_date + datetime.timedelta( days=1 ) #for better code timing, could iterate every 7 days
```
