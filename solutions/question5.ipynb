```python
f = open( '../prompts/5/shrek.txt', 'r')
transcript = f.read()
f.close()
```

## String Cleaning

```python
# turn everything lowercase
transcript_lower = transcript.lower()
```

```python
print ( transcript_lower[:1000] )
```

```python
#get a list of words
words = transcript_lower.split( ' ' )
words = [ word.strip() for word in words ]
```

```python
print ( words[:100] )
```

```python
love_words = [ word for word in words if 'love' in word ]
```

```python
print ( love_words[:100] )
```

```python
import pandas as pd
df_love_words = pd.DataFrame( {'word': love_words} )
```

```python
df_love_words
```

```python
df_love_words['word'].value_counts()
```

```python
def remove_punctuation( word: str ) -> str:

    """given a word, remove the punctuation"""

    punctuation = ['!','?',',','.',']]'] #do not include apostrophe
    for punc in punctuation:
        word = word.replace( punc, '' )

    return word
```

```python
remove_punctuation('love????')
```

```python
df_love_words['stripped'] = df_love_words['word'].apply( lambda word: remove_punctuation(word) )
```

```python
df_love_words
```

```python
df_love_words['stripped'].value_counts()
```

```python
df_love_words['is_love'] = df_love_words['stripped'].isin( ['love',"love's"] )
```
